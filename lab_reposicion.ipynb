```python
# Importacion de las librerias
import pandas as pd
import seaborn as sns
```

```python
# Se imprime la data
data.info()
```

```python
# Se muestra el mapa para conocer los datos vacios
sns.heatmap(data.isnull(), yticklabels = False, cbar = False, cmap = 'viridis')
```

```python
# Dropear los datos con NA
data.dropna(subset = ['education'], inplace = True)
data.dropna(subset = ['cigsPerDay'], inplace = True)
data.dropna(subset = ['BPMeds'], inplace = True)
data.dropna(subset = ['totChol'], inplace = True)
data.dropna(subset = ['BMI'], inplace = True)
data.dropna(subset = ['glucose'], inplace = True)
data.dropna(subset = ['heartRate'], inplace = True)
```

```python
# Se muestra el mapa para conocer los datos vacios
sns.heatmap(data.isnull(), yticklabels = False, cbar = False, cmap = 'viridis')
```

### Task 1.2

```python
# Importacion de librerias
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score
```

```python
# Variables
X = data.iloc[:, :-1]
y = data.iloc[:, -1]
```

```python
# Modelo polinomial
polynomial = PolynomialFeatures(degree=2, include_bias=False)
X_polynomial = polynomial.fit_transform(X)

# Se obtienen los datos de entreno y test
X_train, X_test, y_train, y_test = train_test_split(X_polynomial, y, test_size=0.1, random_state=42)

# Se hace la regresion logistica
logistic_model = LogisticRegression(max_iter=500)
logistic_model.fit(X_train, y_train)
```

```python
prediction = logistic_model.predict(X_test)
confusion_matrix(y_test, prediction)
```

```python
accuracy_score(y_test, prediction)
```

```python
precision_score(y_test, prediction)
```
